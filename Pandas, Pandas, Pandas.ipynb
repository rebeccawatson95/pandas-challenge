```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
#school_data = ("C:/Users/rebec/Desktop/UCI/Week 4/schools_complete-data.csv")
#student_data = ("C:/Users/rebec/Desktop/UCI/Week 4/students_complete-data.csv")
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Combine the data into a single dataset.  
complete_school_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
complete_school_df.head()
```

```python
#DISTRICT SUMMARY
#Create a high-level snapshot, in a DataFrame, of the district's key metrics, including the following:

#Calculate the total number of schools
num_schools = len(complete_school_df['school_name'].unique())
#Calculate the total number of students
num_student = len(complete_school_df['student_name'])
#Calculate the total budget
total_budget = sum(complete_school_df['budget'].unique())
#Calculate the average math score 
avg_math = complete_school_df['math_score'].mean()
#Calculate the average reading score
avg_read = complete_school_df['reading_score'].mean()
#Calculate the percentage of students with a passing math score (70 or greater)
pass_math_df= complete_school_df.loc[(complete_school_df['math_score']>=70),:] 
percent_pass_math = len(pass_math_df)/ num_student * 100
#Calculate the percentage of students with a passing reading score (70 or greater)
pass_read_df= complete_school_df.loc[(complete_school_df['reading_score']>=70),:] 
percent_pass_read = len(pass_read_df)/ num_student * 100
#Calculate the percentage of students who passed math and reading (% Overall Passing)
pass_all_df= complete_school_df.loc[(complete_school_df['reading_score']>=70) & (complete_school_df['math_score']>=70),:] 
percent_pass_all = len(pass_all_df)/ num_student * 100
```

```python
#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools": [num_schools],
                              "Total Students": [num_student],
                              "Total Budget": [total_budget],
                              "Average Math Score": [avg_math],
                                    "Average Reading Score": [avg_read],
                                   "% Passing Math": [percent_pass_math],
                                   "% Passing Reading": [percent_pass_read], 
                                   "% Overall Passing": [percent_pass_all]})
#Optional: give the displayed data cleaner formatting
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.2f}'.format)
district_summary_df
```

```python
#SCHOOL SUMMARY
#Create an overview table that summarizes key metrics about each school, including:
#School Name
school_index = school_df.set_index("school_name")
#School Type
school_type = school_index['type']
#Total Students
school_students = school_index['size']
#Total School Budget
school_budget = school_index['budget']
#Per Student Budget
per_student_budget = school_budget/school_students

index=complete_school_df.set_index("school_name")
byschool=index.groupby(["school_name"])

#Average Math Score
school_avg_math = byschool['math_score'].mean()
#Average Reading Score
school_avg_read = byschool['reading_score'].mean()

#% Passing Math
num_stu_perschool = byschool['student_name'].count()
byschool_math = pass_math_df.groupby("school_name")
byschool_pass_math = byschool_math['student_name'].count()/num_stu_perschool*100

#% Passing Reading
byschool_read = pass_read_df.groupby("school_name")
byschool_pass_read = byschool_read['student_name'].count()/num_stu_perschool*100

#% Overall Passing (The percentage of students that passed math and reading.)
pass_all_byschool_df= complete_school_df.loc[(complete_school_df['reading_score']>=70) & 
                    (complete_school_df['math_score']>=70),:].groupby('school_name')['Student ID'].count()/num_stu_perschool*100 
```

```python
#Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type": school_type,
                              "Total Students": school_students,
                              "Total School Budget": school_budget,
                              "Per Student Budget": per_student_budget,
                                "Average Math Score": school_avg_math,
                                "Average Reading Score": school_avg_read,
                                "% Passing Math": byschool_pass_math,
                                "% Passing Reading": byschool_pass_read,
                                "% Overall Passing": pass_all_byschool_df})
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map('${:,.2f}'.format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map('${:,.2f}'.format)
school_summary_df
```

```python
#TOP PERFORMING SCHOOLS (By % Overall Passing)
#Sort and display the top five performing schools by % overall passing.
top_school_summary = school_summary_df.sort_values("% Overall Passing", ascending=False)
top_school_summary.head()
```

```python
#BOTTOM PERFORMING SCHOOLS (By % Overall Passing)
#Sort and display the five worst-performing schools by % overall passing.
bottom_school_summary = school_summary_df.sort_values("% Overall Passing", ascending=True)
bottom_school_summary.head()
```

```python
#MATH SCORES BY GRADE
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school
ninth = complete_school_df.loc[complete_school_df['grade']=='9th'].groupby("school_name")
tenth = complete_school_df.loc[complete_school_df['grade']=='10th'].groupby("school_name")
eleventh = complete_school_df.loc[complete_school_df['grade']=='11th'].groupby("school_name")
twelfth = complete_school_df.loc[complete_school_df['grade']=='12th'].groupby("school_name")

ninth_math = ninth['math_score'].mean()
tenth_math = tenth['math_score'].mean()
eleventh_math = eleventh['math_score'].mean()
twelfth_math = twelfth['math_score'].mean()

#Combine the series into a dataframe
math_summary = pd.DataFrame({"9th": ninth_math,
                            "10th": tenth_math,
                            "11th": eleventh_math,
                            "12th": twelfth_math})
math_summary = math_summary[["9th","10th","11th","12th"]]
#Optional: give the displayed data cleaner formatting
math_summary.index.name=None

math_summary
```

```python
#READING SCORES BY GRADE 
#Perform the same operations as above for reading scores
ninth = complete_school_df.loc[complete_school_df['grade']=='9th'].groupby("school_name")
tenth = complete_school_df.loc[complete_school_df['grade']=='10th'].groupby("school_name")
eleventh = complete_school_df.loc[complete_school_df['grade']=='11th'].groupby("school_name")
twelfth = complete_school_df.loc[complete_school_df['grade']=='12th'].groupby("school_name")

ninth_read = ninth['reading_score'].mean()
tenth_read = tenth['reading_score'].mean()
eleventh_read = eleventh['reading_score'].mean()
twelfth_read = twelfth['reading_score'].mean()

#Combine the series into a dataframe
read_summary = pd.DataFrame({"9th": ninth_read,
                            "10th": tenth_read,
                            "11th": eleventh_read,
                            "12th": twelfth_read})
read_summary = read_summary[["9th","10th","11th","12th"]]
#Optional: give the displayed data cleaner formatting
read_summary.index.name=None

read_summary
```

```python
#SCORES BY SCHOOL SPENDING
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. 
#Include in the table each of the following:

bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

complete_school_df['spending_ranges'] = pd.cut(complete_school_df['budget']/complete_school_df['size'], bins, labels = group_names, include_lowest=True)
make_groups = complete_school_df.groupby('spending_ranges')

#Average Math Score
avg_math = make_groups['math_score'].mean()
#Average Reading Score
avg_read = make_groups['reading_score'].mean()
#% Passing Math
pass_math = complete_school_df[complete_school_df['math_score'] >= 70].groupby('spending_ranges')['Student ID'].count()/make_groups['Student ID'].count()*100
#% Passing Reading
pass_read = complete_school_df[complete_school_df['reading_score'] >= 70].groupby('spending_ranges')['Student ID'].count()/make_groups['Student ID'].count()*100
#Overall Passing Rate (Average of the above two)
overall = complete_school_df[(complete_school_df['reading_score'] >= 70) & (complete_school_df['math_score'] >= 70)].groupby('spending_ranges')['Student ID'].count()/make_groups['Student ID'].count()*100
 
scores_by_spending = pd.DataFrame({"Average Math Score": avg_math,
                                  "Average Reading Score": avg_read,
                                  "% Passing Math": pass_math,
                                  "% Passing Reading": pass_read,
                                  "% Overall Passing": overall})
scores_by_spending.index.name = "Spending Ranges (Per Student)"
scores_by_spending
```

```python
#SCORES BY SCHOOL SIZE 
#Perform the same operations as above, based on school size.
bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

complete_school_df['size_ranges'] = pd.cut(complete_school_df['size'], bins, labels = group_names, include_lowest=True)
make_groups = complete_school_df.groupby('size_ranges')

#Average Math Score
avg_math = make_groups['math_score'].mean()
#Average Reading Score
avg_read = make_groups['reading_score'].mean()
#% Passing Math
pass_math = complete_school_df[complete_school_df['math_score'] >= 70].groupby('size_ranges')['Student ID'].count()/make_groups['Student ID'].count()*100
#% Passing Reading
pass_read = complete_school_df[complete_school_df['reading_score'] >= 70].groupby('size_ranges')['Student ID'].count()/make_groups['Student ID'].count()*100
#Overall Passing Rate (Average of the above two)
overall = complete_school_df[(complete_school_df['reading_score'] >= 70) & (complete_school_df['math_score'] >= 70)].groupby('size_ranges')['Student ID'].count()/make_groups['Student ID'].count()*100
 
scores_by_size = pd.DataFrame({"Average Math Score": avg_math,
                                  "Average Reading Score": avg_read,
                                  "% Passing Math": pass_math,
                                  "% Passing Reading": pass_read,
                                  "% Overall Passing": overall})
scores_by_size.index.name = "School Size"
scores_by_size
```

```python
#SCORES BY SCHOOL TYPE
#Perform the same operations as above, based on school type
school_type = complete_school_df.groupby("type")

#Average Math Score
avg_math = school_type['math_score'].mean()
#Average Reading Score
avg_read = school_type['reading_score'].mean()
#% Passing Math
pass_math = complete_school_df[complete_school_df['math_score'] >= 70].groupby('type')['Student ID'].count()/school_type['Student ID'].count()*100
#% Passing Reading
pass_read = complete_school_df[complete_school_df['reading_score'] >= 70].groupby('type')['Student ID'].count()/school_type['Student ID'].count()*100
#Overall Passing Rate (Average of the above two)
overall = complete_school_df[(complete_school_df['reading_score'] >= 70) & (complete_school_df['math_score'] >= 70)].groupby('type')['Student ID'].count()/school_type['Student ID'].count()*100
 
scores_by_size = pd.DataFrame({"Average Math Score": avg_math,
                                  "Average Reading Score": avg_read,
                                  "% Passing Math": pass_math,
                                  "% Passing Reading": pass_read,
                                  "% Overall Passing": overall})
scores_by_size.index.name = "School Type"
scores_by_size
```
